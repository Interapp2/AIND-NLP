```python
import os
import requests
import re
from bs4 import BeautifulSoup
```

```python
text_download = requests.get("https://news24.com")
print(text_download.text)
```

```python
parsed_soup = BeautifulSoup(text_download.text, "html5lib")
print(parsed_soup.get_text())
```

```python
headlines = parsed_soup.find_all("div",class_="news_item")
print(headlines[5])
```

```python
headlines[5].select_one("h4 a").get_text().strip()
```

```python
headlines[5].select_one("p").get_text().strip()
```
